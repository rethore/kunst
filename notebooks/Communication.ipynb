We communicate through threads of topics, where people may want to reply to one aspect, but not necessarily to the whole scope of an email.
So we need some kind of subthreading system where a message can be forked into different threads representing different topics.
Similarly a common topic might re-emerge in threads, and we would like to have the context of that discussion coming back. So we need a way to connect together different threads of discussion together and regroup them to existing topics. 

This could be a manual process, through for example tagging a message, to regroup it to an existing discussion thread, but that's kind of ackwards, and not everybody would actually do it. You would need some dedicated souls to regroup the message together through manual tagging.

This is kind of a collective mind-maping, where you would want the individual bubbles to be virtually connected to other mind-maps.

We could potentially use some kind of machine learning process, an undirected graph approach, that find connection between messages, and somehow create those virtual links. One easy approach is to use the words used in the text, so that we can see if other messages are related to that topic. Some kind of genetic score in a sense. The more two messages have the same words, or series of words, the more their DNA relationship score is high. 

The context of the thread might also play a role, for instance if a cell has a high score to relate it to another thread of discussion, then the parents and childs of the cell should also have a relatively higher discussion score.

So instead of having some strict categories, we would have some very loose relationship between discussion that would traverse discussions.

So two problems here:

* How do we effectively represent the discussion process in both directions: 
    * The natural dynamical iterative discussion
    * The context of the current discussion
    * The topic of discussion that might include several similar parallel discussions

So in a way it's like having a direct context and an indirect context that might be unkown from the person discussing

We could use some kind of reddit type of indentation for the direct context, but the indirect context is less easy to express. You could have a "related discussions" in a parallel frame, or under the discussion. It might be easier to have to click/touch the discussion to orient the lecture direction. On a phone screen there isn't much space horizontally, so it's a vertical navigation.

The alternative discussions, could be indicated as collapsed discussion threads. By clicking on it the original discussion thread is collapsed, and the new one is extended.

Alternatively the discussions could be linked manually, each discussion cell has a unique ID, which could be referenced in the metadata as a list of directly related cells. There could also be an indirectly manual process, where people could themselves validate/invalidate a proposed connection, by clicking on a connection button manually.
